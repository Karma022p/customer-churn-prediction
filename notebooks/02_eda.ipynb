```python
# Act4: EDA
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("customer_churn_cleaned.csv")
df.head()
```

```python
# Basic summary
print(df.shape)
print(df.info())
print(df["Churn"].value_counts(normalize=True))
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df["tenure"], bins=30)
plt.title("Tenure Distribution")
plt.xlabel("Tenure (months)")
plt.tight_layout()
plt.savefig("tenure_distribution.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df["MonthlyCharges"])
plt.title("Monthly Charges Distribution")
plt.tight_layout()
plt.savefig("monthly_charges_boxplot.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x="Contract", hue="Churn", data=df)
plt.title("Contract Type vs Churn")
plt.tight_layout()
plt.savefig("contract_vs_churn.png", dpi=300)
plt.show()
```

```python
# small numeric correlation to show relationships
num_cols = ["tenure","MonthlyCharges","TotalCharges","SeniorCitizen","Churn"]
plt.figure(figsize=(6,4))
sns.heatmap(df[num_cols].corr(), annot=True, fmt=".2f")
plt.title("Numeric Correlation (small subset)")
plt.tight_layout()
plt.savefig("correlation_small.png", dpi=300)
plt.show()
```
